```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.shape
```

# Barplot

```python
sns.barplot(x="sex",y="total_bill",data=tips)
```

```python
import numpy as np
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.std)
```

```python
sns.countplot(x="time",data=tips)
```

# BoxPlot

```python
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker")
```

# Violinplot

```python
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
#use of split =True
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)
```

```python
sns.violinplot(x="day",y="tip",data=tips,hue="smoker",split=True)
```

# Distribution plots

### distplot

#### it is univariate

```python
sns.distplot(tips["total_bill"])    # this line is called kde(kernel density estimation)
```

### kdeplot

```python
sns.kdeplot(tips["total_bill"])
```

### jointplot

#### it has two variables i.e. we can combine the two different distribution plots into one

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
```

#### we can see that there is a linear trend between the total_bill and tip

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
sns.jointplot(x="total_bill",y="size",data=tips,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="size",data=tips,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="size",data=tips,kind="scatter")
```
